```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\PROJECT\data\processed\clean_survey.csv")

print(df['treatment'].value_counts(dropna=False))
print(df['work_interfere'].value_counts(dropna=False))
```

```python
# Observation:
# Created a binary target column at_risk (1 = Yes, 0 = No).
# Distribution plot shows a slight class imbalance, with more respondents at risk (1) compared to not at risk (0).


df['at_risk'] = df['treatment'].map({'Yes': 1, 'No': 0})
print(df['at_risk'].value_counts(normalize=True))
sns.countplot(x='at_risk', data=df)
plt.title("Target distribution (0 = not at risk, 1 = at risk)")
plt.show()
```

```python
df.to_csv(r"C:\Users\HP\OneDrive\Desktop\PROJECT\data\processed\target_ready.csv", index=False)
print("✅ Saved dataset with target column added.")
```
